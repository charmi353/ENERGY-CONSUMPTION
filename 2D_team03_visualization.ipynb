```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Energy_consumption_dataset_uncleaned.csv")
hvac_sum=df["HVACUsage"].sum()
lighting_sum=df["LightingUsage"].sum()
renewable_sum=df["RenewableEnergy"].sum()
values=[hvac_sum,lighting_sum,renewable_sum]
labels=["HVACUsage","LightingUsage","RenewableEnergy"]
plt.pie(values,labels=labels,autopct="%1.1f%%")
plt.title("Energy Usage Distribution")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv("Energy_consumption_dataset_uncleaned.csv")
values=[df["HVACUsage"].sum(),df["LightingUsage"].sum(),df["RenewableEnergy"].sum()]
labels=["HVACUsage","LightingUsage","RenewableEnergy"]
total = sum(values)
percentages = [int(value / total * 100) for value in values]
blocks=[]
percentages[-1] = 100 - sum(percentages[:-1])
for i in range(len(percentages)):
    blocks +=[i]*percentages[i]
blocks=np.array(blocks)
blocks=blocks.reshape(10,10)
plt.imshow(blocks)
plt.xticks([])
plt.yticks([])
plt.title("Energy Usage waffle chart")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Energy_consumption_dataset_uncleaned.csv")
from wordcloud import WordCloud
df["Category"]=pd.cut(df["EnergyConsumption"],bins=3,labels=["Low","Medium","High"])
text=" ".join(df["Category"].astype(str))
wc=WordCloud(background_color="white").generate(text)
plt.imshow(wc)
plt.axis("off")
plt.show()
```
