```python
import pandas as pd
df=pd.read_csv("Energy_consumption_dataset_uncleaned.csv")
nums_cols=df.select_dtypes(include=['int64','float64'])
outlier_count={}
for col in nums_cols:
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    lb=q1-1.5*iqr
    ub=q3+1.5*iqr
    outliers=df[(df[col]<lb)|(df[col]>ub)]
    outlier_count[col]=len(outliers)
print(outlier_count)
df.to_csv("Energy_consumption_dataset_uncleaned.csv",index=False)
```

```python
import pandas as pd
df=pd.read_csv("Energy_consumption_dataset_uncleaned.csv")
numeric_cols=["Month","Hour","Temperature","Humidity","SquareFootage","Occupancy","RenewableEnergy","EnergyConsumption"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
c_cols=["DayOfWeek","Holiday","HVACUsage","LightingUsage"]
for col in c_cols:
    df[col] = df[col].str.strip().str.lower()
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
print(df.head(10))
df.to_csv("Energy_consumption_dataset_uncleaned.csv",index=False)

```

```python
import pandas as pd
df=pd.read_csv("Energy_consumption_dataset_uncleaned.csv")
df["DayOfWeek"] = df["DayOfWeek"].astype("category").cat.codes
df["Holiday"] = df["Holiday"].map({"yes": 1, "no": 0})
df["HVACUsage"] = df["HVACUsage"].map({"on": 1, "off": 0})
df["LightingUsage"] = df["LightingUsage"].map({"on": 1, "off": 0})
df["EnergyConsumption_binned"]=df["EnergyConsumption_binned"].map({"low":0,"medium":1,"high":2})
numeric_cols=["Month","Hour","DayOfWeek","Holiday","Temperature","Humidity","SquareFootage","Occupancy","HVACUsage","LightingUsage","RenewableEnergy","EnergyConsumption","EnergyConsumption_binned"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
df[numeric_cols] = (df[numeric_cols] - df[numeric_cols].min()) / (df[numeric_cols].max() - df[numeric_cols].min())
print(df.head(10))
#df.to_csv("Energy_consumption_dataset_uncleaned.csv",index=False)
```
